````python
import numpy as np
from ripser import ripser
from ripser import Rips
from persim import plot_diagrams
import matplotlib.pyplot as plt
from pprint import pprint
````

#### How to call ripser
```ripser.ripser(X, maxdim=1, thresh=inf, coeff=2, distance_matrix=False, do_cocycles=False, metric='euclidean', n_perm=None)```

[Docs](https://ripser.scikit-tda.org/en/latest/reference/stubs/ripser.ripser.html#ripser.ripser)

````python
data = np.random.random((100, 2))
diagrams = ripser(data)['dgms']
print(ripser(data).keys())
plot_diagrams(diagrams, show=True)
````

#### How to call ripser
```class ripser.Rips(maxdim=1, thresh=inf, coeff=2, do_cocycles=False, n_perm=None, verbose=True)```

[Docs](https://ripser.scikit-tda.org/en/latest/reference/stubs/ripser.Rips.html#ripser.Rips)

````python
rips = Rips()
data = np.random.random((100,2))
diagrams = rips.fit_transform(data)
rips.complex_plot(diagrams)
````

````python
data = np.array([(5, 5), (10, 10), (10, 5), (5, 10),
                 (-5, -5), (-10, -10), (-10, -5), (-5, -10)])
plt.scatter(data[:,0], data[:,1], color='b')

ripser_res = ripser(data)
diagrams = ripser_res['dgms']
pprint(ripser_res)
plt.scatter(diagrams[0][:,0], diagrams[0][:,1], color='r')
# plt.complex_plot(x[edges.T], y[edges.T], linestyle='-', color='y',
#         markerfacecolor='red', marker='o') 
plt.show()
plot_diagrams(diagrams, show=True)
````

### Using the homology package

````python
from homology import homology

cc = homology.ChainComplex()

data = [(5,10), (10,10), (7.5, 5)]
for ind, group in enumerate(data):
    cc.add_group(ind, 2)
    cc.add_factor(ind, [group])
````

````python
[print(cc.get_homology(i)) for i in range(len(data))]
````

````python
print(cc.generate_cochain_complex())
````

### moguTDA: Python package for Simplicial Complex
- [Article about MoguTDA](https://medium.com/@stephenhky/mogutda-python-package-for-simplicial-complex-60b87da99712)
- [MoguTDA Repo](https://github.com/stephenhky/MoguTDA)

````python
import numpy as np
from mogutda import SimplicialComplex
````

````python

squares_points = np.array([(5,5), (10,10), (10, 5), (5, 10), 
                            (-5, -5), (-10,-10), (-10, -5), (-5, -10)])
plt.scatter(squares_points[:,0], squares_points[:,1], color='c')
plt.title('Triangle Data Points')
plt.axis('equal')
plt.show()
squares_sc = SimplicialComplex(simplices=squares_points)
````

````python
print(squares_sc.betti_number(0))
print(squares_sc.betti_number(1))
print(squares_sc.betti_number(2))
````

````python
squares_sc.draw()
````

````python
torus_points = np.array([(1,2,4), (4,2,5), (2,3,5), (3,5,6), (5,6,1), (1,6,2), (6,7,2), (7,3,2),
                (1,3,4), (3,4,6), (4,6,7), (4,5,7), (5,7,1), (7,3,1)])
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(torus_points[:,0], torus_points[:,1], torus_points[:, 2], color='c', marker='^')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
````

````python
torus_sc = SimplicialComplex(simplices=torus_points)

print("Simplices: ", torus_sc.simplices)
print("Face Set (type: {}): {}".format(type(torus_sc.face_set), torus_sc.face_set))
print("Betti Number 1: ", torus_sc.betti_number(0))
print("Betti Number 2: ", torus_sc.betti_number(1))
print("Betti Number 3: ", torus_sc.betti_number(2))
torus_sc.draw()
````
